LIMPIEZA DE DATOS

```python
df_historico_partidos = pd.read_csv('./../Data/partidos_champions_data.csv')
df_historico_partidos[df_historico_partidos['season']==1969].shape
```

```python
#ELiminamos la columna Unnamed:0
df_historico_partidos.drop(['Unnamed: 0'], axis=1, inplace=True)
df_historico_partidos.rename(columns={'Home': 'HomeTeam', 'Away':'AwayTeam','season': 'Season'}, inplace=True)
```

```python
df_historico_partidos
```

```python
df_historico_partidos[df_historico_partidos['GoalHome'].str.contains('^\D')]
```

En este caso, tenemos los partidos que se decidieron por penaltis (el resultado viene entre paréntesis) y los partidos que no se llegaron a disputar (vienen entre guiones). Vamos a eliminar los partidos que no se llegaron a disputar.

```python
df_partidos_suspendidos = df_historico_partidos[df_historico_partidos['GoalAway'].str.contains('-')]
df_partidos_suspendidos.shape
```

```python
df_historico_partidos.drop(df_partidos_suspendidos.index, inplace = True)
df_historico_partidos.shape
```

Como podemos observar, se han eliminado los 18 partidos que estaban suspendidos

```python
df_historico_partidos[df_historico_partidos['GoalHome'].str.contains('-')]
```

```python
df_historico_partidos[df_historico_partidos['GoalAway'].str.contains('-')]
```

```python
df_historico_partidos.dtypes
```

```python
df_penaltis = df_historico_partidos[df_historico_partidos['GoalHome'].str.contains('[^\d]')]
df_penaltis
```

```python
df_penaltis.shape
```

```python
df_historico_partidos.drop(df_penaltis.index, inplace=True)
df_historico_partidos.shape
```

```python
df_historico_partidos= df_historico_partidos.astype({
                                    'GoalHome':int,
                                    'GoalAway':int
                                    }
                               )
```

```python
df_historico_partidos.dtypes
```

```python
df_penaltis['GoalAway']=df_penaltis['GoalAway'].str.replace('[^\d]', '', regex=True)
df_penaltis['GoalHome']=df_penaltis['GoalHome'].str.replace('[^\d]', '', regex=True)
```

```python
df_penaltis
```

```python
df_historico_partidos['HomeTeam'] = df_historico_partidos['HomeTeam'].str.strip()
df_historico_partidos['AwayTeam'] = df_historico_partidos['AwayTeam'].str.strip()

```
